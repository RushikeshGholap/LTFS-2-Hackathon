```python
def preprocess(df):
    df = df.fillna(0)
    for i in range(len(df)): 
        df.loc[i,'Day'] = int(df.loc[i,'application_date'].split('-')[2])
        df.loc[i,'month'] = int(df.loc[i,'application_date'].split('-')[1])
        df.loc[i,'year'] = int(df.loc[i,'application_date'].split('-')[0])

    train_dummies  = pd.get_dummies(df['state'], prefix = 'state')
    df = pd.concat([df, train_dummies], axis=1)
    train_dummies  = pd.get_dummies(df['zone'], prefix = 'zone')
    df = pd.concat([df, train_dummies], axis=1)
    df.drop(columns=['application_date','state','zone'],inplace=True)    
    
    return(df)
```

```python
train_pre = preprocess(train)
```

```python
train_pre.info()
```

```python
X = train_pre.drop("case_count",1)
y = train_pre['case_count']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=7)
```

```python
lr = RandomForestRegressor()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
print(metrics.mean_absolute_error(y_test,y_pred))
```
